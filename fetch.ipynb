移動平均線を作ってみる

```python
ticker="AMZN"
data = yf.download(ticker, start="2022-01-01", end="2024-12-31", interval="1d")
# マルチインデックスカラムの場合
if isinstance(data.columns, pd.MultiIndex):
    # 第1レベル（Price, Close, High, Low, Open）だけをカラムとして使用
    data.columns = data.columns.get_level_values(0)

price = data["Close"]
span = 30

data["MA5d"] = price.rolling(window=span,min_periods=1).mean()
```

```python
df
```

```python
data_ma = data[["Close", "MA5d"]]
data_ma
```

```python
# plt.plotで折れ線グラフを描画
plt.figure(figsize=(15, 5))
plt.plot(data_ma.index, data_ma)
plt.title("株価の推移")
plt.xlabel("date")
plt.ylabel("price")
plt.legend(data_ma)
plt.show()
```

```python
data["std"] = price.rolling(window=span,min_periods=1).std()
data
```

```python
# ボリンジャーバンドの作成と可視化
data["-1o"] = data["MA5d"] - data["std"]
data["+1o"] = data["MA5d"] + data["std"]
data["-2o"] = data["MA5d"] - 2 * data["std"]
data["+2o"] = data["MA5d"] + 2 * data["std"]

plt.figure(figsize=(15, 5))
plt.title("trend analysis")
plt.plot(data["Close"], color="blue", label="close")
plt.plot(data["-1o"], color="red", label="-1o")
plt.plot(data["+1o"], color="red", label="+1o")
plt.fill_between(data.index, data["-1o"], data["+1o"], color="red", alpha=0.5)
plt.plot(data["-2o"], color="lightgreen", label="-2o")
plt.plot(data["+2o"], color="lightgreen", label="+2o")
plt.fill_between(data.index, data["-2o"], data["+2o"], color="lightgreen", alpha=0.5)
plt.xlabel("date")
plt.ylabel("Close price")
plt.legend(["close", "-1o", "+1o"], loc="upper left")
plt.show()
```

終値がボリンジャーバンドを突き抜けて上がりそうな時に購入すると、大幅な上昇が期待できる
